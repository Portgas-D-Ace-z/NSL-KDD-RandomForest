# Scaling

```python
from sklearn.preprocessing import StandardScaler
#train
x_train_num = x_train_data_encoded.select_dtypes(include=[np.number])
x_test_num = x_test_data_encoded.select_dtypes(include=[np.number])

std_scaler = StandardScaler()

x_train_scaled = std_scaler.fit_transform(x_train_num)
x_test_scaled = std_scaler.transform(x_test_num)

#test
df_test_num = df_test_data_encoded.select_dtypes(include=[np.number])
df_test_scaled = std_scaler.transform(df_test_num)
```

# Train

```python
from sklearn.ensemble import RandomForestClassifier
#train


r_forest_clf=RandomForestClassifier(random_state=42)
r_forest_clf.fit(x_train_scaled, y_train)

y_pred = r_forest_clf.predict(x_test_scaled)
y_pred

#test
df_pred = r_forest_clf.predict(df_test_scaled)
```

```python
y_train.isna().sum()
```

# Fine tune

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

#train
param_distribs = {"max_depth": randint(low=1, high=50), 
    "max_features": randint(low=1, high=100), 
    "min_samples_split": randint(low=2, high=50), 
    "bootstrap": [True, False], 
    "criterion": ["gini", "entropy","log_loss"],
    "n_estimators": randint(low=10,high=100) ,}

rnd_search = RandomizedSearchCV(r_forest_clf, param_distributions=param_distribs, 
    n_iter=50, cv=3,random_state=42) 
rnd_search.fit(x_train_scaled, y_train)
y_pred_rand = rnd_search.predict(x_test_scaled)
y_pred_rand

#test
df_pred_rand = rnd_search.predict(df_test_scaled)
```
